## Project Overview
This notebook performs sentiment analysis on the Amazon Fine Food Reviews dataset. We will clean the data, calculate sentiment scores using TextBlob, and visualize the findings.

### Task 1 — Data Loading & Exploration
Load the CSV file using Pandas and explore its structure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import os

# File path
file_path = 'Reviews.csv'

# Check if file exists, if not create a sample for demonstration
if not os.path.exists(file_path):
    print("Reviews.csv not found. Creating a sample dataset for demonstration...")
    sample_data = {
        'Text': [
            'This product is amazing! I love it.',
            'Awful experience. The food was stale.',
            'It was okay, nothing special.',
            'Great taste and fast delivery.',
            'Very disappointed. Will not buy again.',
            'Delicious! My whole family liked it.',
            'Not bad, but could be better.',
            'Horrible taste. Avoid at all costs.',
            'Absolutely wonderful. High quality.',
            'Meh, it was mediocre.'
        ],
        'Score': [5, 1, 3, 5, 1, 5, 3, 1, 5, 2]
    }
    pd.DataFrame(sample_data).to_csv(file_path, index=False)

# Load the dataset (first 5,000 rows as requested)
# We use error_bad_lines=False or on_bad_lines='skip' to handle formatting issues in CSV
try:
    df = pd.read_csv(file_path, nrows=5000, on_bad_lines='skip')
except Exception as e:
    print(f"Error loading CSV: {e}")
    df = pd.DataFrame() # Fallback

if not df.empty:
    # Display the first 10 rows
    print("First 10 rows:")
    display(df.head(10))

    # Check dimensions
    print(f"\nDataset Shape: {df.shape}")

    # Identify columns
    print(f"\nColumns: {df.columns.tolist()}")
```

### Task 2 — Data Cleaning
Remove nulls, duplicates, and keep relevant columns.

```python
if not df.empty:
    # Check for null values
    print("Null values before cleaning:")
    print(df.isnull().sum())

    # Remove rows where review text is empty
    # In Kaggle dataset, column name is 'Text'
    text_col = 'Text' if 'Text' in df.columns else df.columns[0]
    score_col = 'Score' if 'Score' in df.columns else df.columns[1]
    
    df.dropna(subset=[text_col], inplace=True)

    # Remove duplicate reviews
    df.drop_duplicates(subset=[text_col], inplace=True)

    # Keep only necessary columns: review text + rating (score)
    df = df[[text_col, score_col]]

    print(f"\nDataset shape after cleaning: {df.shape}")
```

### Task 3 — Sentiment Analysis
Calculate polarity score and label sentiments.

```python
if not df.empty:
    def get_sentiment(text):
        analysis = TextBlob(str(text))
        return analysis.sentiment.polarity

    # Calculate polarity score
    df['Polarity'] = df[text_col].apply(get_sentiment)

    # Label sentiments
    def label_sentiment(score):
        if score > 0:
            return 'Positive'
        elif score < 0:
            return 'Negative'
        else:
            return 'Neutral'

    df['Sentiment'] = df['Polarity'].apply(label_sentiment)

    display(df.head())
    print("\nSentiment counts:")
    print(df['Sentiment'].value_counts())
```

### Task 4 — Visualization
Generate charts to represent the data.

```python
if not df.empty:
    # Set visual style
    sns.set_theme(style="whitegrid")

    # Chart 1: Bar chart showing count of sentiments
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Sentiment', data=df, palette='viridis', hue='Sentiment', legend=False)
    plt.title('Distribution of Sentiments')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.show()

    # Chart 2: Pie chart showing percentage distribution
    plt.figure(figsize=(8, 8))
    df['Sentiment'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)
    plt.title('Percentage Distribution of Sentiments')
    plt.ylabel('')
    plt.show()

    # Chart 3: Rating (Score) vs Sentiment Polarity
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=score_col, y='Polarity', data=df, palette='coolwarm', hue=score_col, legend=False)
    plt.title('Rating vs Sentiment Polarity')
    plt.xlabel('Rating (Score)')
    plt.ylabel('Polarity Score')
    plt.show()
```
